```python
%matplotlib inline
import sys
sys.path.append("..")
import cv2
import util
import matplotlib
import pylab as plt
import numpy as np
import h5py
import os

os.getcwd()
```

```python
h5 = h5py.File("./dataset/val_generated.h5", 'r')
dt = h5["label"]["0000000"]  # try 0000045
labels = np.reshape(dt, (41, 46, 46))
#print(heat_result.shape)
#print(key)

dta = h5["data"]["0000000"]
print(dta.shape)
dta_img = np.reshape(dta, (3, 368, 368))
dta_img0 = np.reshape(dta, (3, 368, 368))
print(dta_img.shape)

dta_img = np.transpose(dta_img, (1, 2, 0))
print(dta_img.shape)
plt.imshow(dta_img[:,:,[2,1,0]])
```

```python
idx = 40  # idx-26
label = labels[idx,:,:]
imageToTest = cv2.resize(dta_img, (0,0), fx=1, fy=1, interpolation=cv2.INTER_CUBIC)
imageToTest_padded, pad = util.padRightDownCorner(imageToTest, 8, 128)

heatmap = cv2.resize(label, (0,0), fx=8, fy=8, interpolation=cv2.INTER_CUBIC)
heatmap = heatmap[:imageToTest_padded.shape[0]-pad[2], :imageToTest_padded.shape[1]-pad[3]]
heatmap = cv2.resize(heatmap, (dta_img.shape[0], dta_img.shape[1]), interpolation=cv2.INTER_CUBIC)

plt.imshow(dta_img[:,:,[2,1,0]])
plt.imshow(heatmap[:,:], alpha=.5)  # red means that vec is negative
```
